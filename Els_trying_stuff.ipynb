```python
import numpy as np
import pandas as pd
from datetime import timedelta
import matplotlib.pyplot as plt

nrows = 481
time_step_seconds = 360
time_start_seconds = 310

df = pd.read_excel('Ilses_biotek_exp.xlsx', skiprows=30, nrows=nrows, usecols="B:CU")

seconds_since_start = np.arange(nrows)*time_step_seconds+time_start_seconds
time_since_start = pd.to_timedelta(seconds_since_start, unit='s')

df['Time'] = time_since_start
df.head()
```

```python
df.set_index('Time', inplace=True)
df.head()
```

```python
fig, ax = plt.subplots()
df.C11.plot()
plt.xlabel(r'timestep', size=15)
plt.ylabel(r'OD600', size=15)
plt.title('growth curve of C11', size=20)
```

```python
fig, ax = plt.subplots()
df.C11.plot()
ax.set_yscale('log')
plt.xlabel(r'timestep', size=15)
plt.ylabel(r'OD600', size=15)
plt.title('growth curve of C11', size=20)
```

```python
logC11 = np.log10(df.C11)


from scipy.stats import linregress

slopeC11, interceptC11, rvalueC11, pvalueC11, blablaC11 = linregress(seconds_since_start[30:200], logC11[30:200])

print(slopeC11, interceptC11, rvalueC11)
```

```python
def modelC11(x):
    return interceptC11 + slopeC11*x


fig, ax = plt.subplots()
df.C11.plot()
plt.plot(seconds_since_start, modelC11(seconds_since_start))

#ax.set_yscale('log')
plt.xlabel(r'timestep', size=15)
plt.ylabel(r'OD600', size=15)
plt.title('growth curve of C11', size=20)
```
